# Exercício 2

image.png

### Freq relativa

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados = { 
    'País':['Estados Unidos', 'Reino Unido','Alemanha','França','Suécia','Suiça','Demais'],
    'freq':[270,101,76,49,30,22,130]  
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
df['freq_rel'] = df.freq.apply(lambda x: round(x/df.freq.sum(),4))
df["freq_rel"]
```

```python
df["freq_rel"].sum() # Valor próximo a 1, indica que os valores estão dentro do esperado.
```

### Gráfico

```python
sns.set_style('darkgrid')

plt.figure(figsize=[12,8])
ax = sns.barplot(x = 'País', y = 'freq', data = df)
ax.set_title('Frequência de ganhadores', loc='left', fontsize=18)
ax.set_xlabel('País'.upper(), fontsize=14)
ax.set_ylabel('Freq'.upper(), fontsize=14)
```

# Exercício 3

image.png

### Frequencia

```python
dados_exercicio_2 = {
    'Sabor':['Chocolate','Napolitano','Flocos','Coco','Morango','Manga'],
    'freq._rel':[0.1,0.185,0.260,0.150,0.080,0.225]
}
```

```python
df_sorvetes = pd.DataFrame(dados_exercicio_2)
df_sorvetes
```

```python
df_sorvetes['freq'] = df_sorvetes['freq._rel'].apply(lambda x: x*200) # 200 pessoas consultadas, basta multiplicar
df_sorvetes
```

```python
df_sorvetes['freq'].sum()
```

### Grafico

```python
sns.set_style('darkgrid')

plt.figure(figsize=[12,8])
ax2 = sns.barplot(x = 'freq', y = 'Sabor', data = df_sorvetes, orient='h')
ax2.set_title('Frequência de sabores', loc='center', fontsize=18)
ax2.set_xlabel('Freq'.upper(), fontsize=14)
ax2.set_ylabel('Sabor'.upper(), fontsize=14)
```

# Exercício 4

image.png

```python
import numpy as np
```

```python
dados_exercicio_4 = {
    'Resposta':['Excelente','Muito Boa','Boa','Razoavel','Ruim'],
    'freq':[50,152,255,np.nan,np.nan],
    'freq._rel': [np.nan,np.nan,np.nan,0.4250,0.1180]
}

dados_exercicio_4
```

```python
df_lojas = pd.DataFrame(dados_exercicio_4)
df_lojas
```

```python
df_lojas_calculos = df_lojas.drop('Resposta',axis=1).copy()
df_lojas_calculos['freq._rel'] = df_lojas['freq'].apply(lambda x : x/1000)
df_lojas_calculos['freq'] = df_lojas['freq._rel'].apply(lambda x : x*1000)

df_lojas = df_lojas.combine_first(df_lojas_calculos)
df_lojas
```

```python
print(df_lojas['freq._rel'].sum())
print(df_lojas['freq'].sum())
```
