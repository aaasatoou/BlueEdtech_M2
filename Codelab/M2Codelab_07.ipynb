### Representa√ß√£o Gr√°fica

Survived - A vari√°vel √© o *target* do dataset, ela ilustra os sobreviventes, ent√£o, √© uma vari√°vel que mostra informa√ß√µes importantes sobre a situa√ß√£o  
Age - a vari√°vel nos permite analisar a media das idades das pessoas no navio e tentar estabelecer alguma correla√ß√£o entre idades e sobreviventes

```python
dados = df_titanic.groupby('Age')['PassengerId'].count().reset_index().rename(columns={"PassengerId":'freq'})

sns.set_style('darkgrid')

plt.figure(figsize=[12,8])
plt.hist(df_titanic['Age'], bins = 8)
plt.title('Grafico de barras', loc='left', fontsize=18)
plt.xlabel('Idades', fontsize=14)
plt.ylabel('Contagem', fontsize=14)
```

```python
dados = df_titanic.groupby('Survived')['PassengerId'].count().reset_index().rename(columns={'PassengerId':'freq'})

sns.set_style('darkgrid')

plt.figure(figsize=[12,8])
sns.barplot(x="Survived", y="freq", data=dados)
plt.title('Grafico de barras', loc='left', fontsize=18)
plt.xlabel('Survived', fontsize=14)
plt.ylabel('Count', fontsize=14)
```

```python
plt.figure(figsize=[12,8])
plt.pie(dados['freq'], labels = ['Morreu', 'Sobreviveu'], autopct='%.0f%%')
plt.show()
```
